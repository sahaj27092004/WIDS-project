```python
import yfinance as yf
import pandas as pd
import numpy as np
import yahoofinancials as YahooFinancials
```

```python
aapl_df = yf.download('AAPL', period="5y")
goog_df = yf.download('GOOG', period="5y")
msft_df = yf.download('MSFT', period="5y")
amzn_df = yf.download('AMZN', period="5y")
```

```python
aapl_df.to_csv("aapl.csv")
goog_df.to_csv("goog.csv")
msft_df.to_csv("msft.csv")
amzn_df.to_csv("amzn.csv")
```
